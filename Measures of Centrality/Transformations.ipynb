In this notebook, i will describe how tranformation(scaling,adding) effects 

1. Mean
2. Median
3. Mode

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
simple_dataset = [x for x in range(1,15,1)]
```

## Mean

```python
def plot_fn(list_to_mode_mean_median,type_of_centrality,legend_value):
    plt.plot(list_to_mode_mean_median,list_to_mode_mean_median,color="gray",marker="o")
    plt.scatter(type_of_centrality,type_of_centrality,c="red",marker="D")
    plt.xlabel("Data Points")
    plt.ylabel("Data Points")
    plt.legend(["Data Points",legend_value])
    print(legend_value + ":",type_of_centrality)
    plt.show()
```

```python
mean = np.mean(simple_dataset)
plot_fn(simple_dataset,mean,"Mean Value")
```

```python
#Scale each element of the list by 4 and then add 3. (Ex: data_point1 = 4*data_point1 + 3) which of the form x = ax + c

Transformed_simple_dataset = [4*x + 3 for x in simple_dataset]
Transformed_simple_dataset
```

```python
Transformed_mean = np.mean(Transformed_simple_dataset)
plot_fn(Transformed_simple_dataset,Transformed_mean,"Transformed Mean Value")
```

Mean for a given matrix X or list of N($x_{1},x_{2},x_{3},...,x_{N}$) data points is given by:

$\bar{X} = 1/N(\sum\limits _{i=1} ^{N}x_{i})$


Now to multiply eaxh element by a and add by c:

$\bar{X'} = 1/N(\sum\limits _{i=1} ^{N}ax_{i} + c)$

$\bar{X'} = 1/N(\sum\limits _{i=1} ^{N}ax_{i}) + 1/N(\sum\limits _{i=1} ^{N}c)$

Since a is constant, we can take it out of summation

$\bar{X'} = a/N(\sum\limits _{i=1} ^{N}x_{i}) + 1/N(\sum\limits _{i=1} ^{N}c)$

$\bar{X'} = a\bar{X} + 1/N(\sum\limits _{i=1} ^{N}c)$

$\bar{X'} = a\bar{X} + 1/N(Nc)$

So, finally 

**$\bar{X'} = a\bar{X} + c$**

Let us check if the above formula is correct with the data we have:

a = 4, c= 3, $\bar{X}$ =7.5,$\bar{X'}$=33

33 = 4(7.5) + 3

LHS = RHS

So, our derivation is correct. 

## Median

```python
Median = np.median(simple_dataset)
plot_fn(simple_dataset,Median,"median")
```

```python
Transformed_median = np.median(Transformed_simple_dataset)
plot_fn(Transformed_simple_dataset,Transformed_median,"Transformed Median")
```

Similarly as mean, The median value gets updated as per the equation:

$\bar Median = aMedian + c$

However the position of median will not be changed.

## Mode

```python
from scipy import stats
```

```python
simple_dataset.append(8)
Mode = int(stats.mode(simple_dataset)[0])
plot_fn(simple_dataset,Mode,"Mode")
```

```python
Transformed_simple_dataset = [4*x + 3 for x in simple_dataset]
Transformed_simple_dataset
```

```python
Transformed_Mode = int(stats.mode(Transformed_simple_dataset)[0])
plot_fn(Transformed_simple_dataset,Transformed_Mode,"Mode")
```

Similarly as mean & median, The mode value gets updated as per the equation:

$\bar Mode = aMode + c$
